# Data cleaning

```python
df = df.dropna(subset=['Attrition'])    # drop the rows in attrition status
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
df.isnull().sum()
```

##  Data Analysis

```python
# Trend in Nurse turnover,demographics and job satisfication

# Attrition Rate overview


Attrition_rate = df['Attrition'].value_counts(normalize=True)
Attrition_rate
```

```python
# visualize Attrition Rate

sns.countplot(x='Attrition',data=df)
plt.title('Attrition Rate')
plt.show()
```

```python
# Analysing trends based on demographics and other factors


# analyse attrition by gender,department,educationlevel

plt.figure(figsize=(10,6))
sns.countplot(x='Gender',hue='Attrition',data=df)
plt.title("Attrition by Gender")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title('Attrition by Job Satisfaction Level')
plt.show()
```
